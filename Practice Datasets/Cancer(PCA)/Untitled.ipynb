```python
df_new = df.drop(['ID','Class'],axis=1)
```

```python
df_new.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std_scaler = StandardScaler()
```

```python
col_names = list(df_new.columns)
```

```python
features = df_new[col_names]
```

```python
features = std_scaler.fit_transform(features.values)
```

```python
scaled_df = pd.DataFrame(features,columns=col_names)
```

```python
scaled_df.head()
```

```python
scaled_df.boxplot(figsize=(20,3))
```

```python
#PCA 
#Step-1 : Create covariance matrix
cov_matrix = np.cov(scaled_df.T)
cov_matrix
```

```python
#Step-2 : Get eigen values and eigen vectors
eig_vals, eig_vecs = np.linalg.eig(cov_matrix)
```

```python
eig_vals
```

```python
eig_vecs
```

```python
tot = sum(eig_vals)
var_exp = [(i/tot)*100 for i in sorted(eig_vals,reverse=True)]
cum_var_exp = np.cumsum(var_exp)
cum_var_exp
```

```python
plt.plot(var_exp)
```

```python
#Using PCA in sklearn
from sklearn.decomposition import PCA
```

```python
#Reducing dimensions from 9 to 7
pca = PCA(n_components=7)
data_reduced = pca.fit_transform(scaled_df)
data_reduced.transpose()
```

```python
pca.components_
```

### Find relations between features and components

```python
df_comp = pd.DataFrame(pca.components_,columns=list(df_new))
```

```python
df_comp.head()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='plasma')
```
