# Normal Distribution Examples

```python
#mean salary = 7,00,000
#std = 90,000
#normal distribution graph
mean = 700000
std = 90000
```

```python
#proba that salary is more than 10,00,000
z1 = (1000000 - mean)/std
z1
```

```python
1 - stats.norm.cdf(z1)
```

```python
#prob that salary is between 6,00,000 and 9,00,000
z2 = (600000-mean)/std
z3 = (900000-mean)/std

stats.norm.cdf(z3) - stats.norm.cdf(z2)
```

```python
#prob that salary is less than 4,00,000
z4 = (400000-mean)/std
stats.norm.cdf(z4)
```

# Poisson Distribution Examples

## Problem Statement 1 - 
If on an average, 6 customers arrive at every 1 min in a bank during busy hours of working:

a) What is probablity that exactly 4 customers arrive in a given 1 min
b) What is probablity that more that 3 customers arrive in a given min

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rate = 6
```

```python
n = np.arange(0,50)
```

```python
poisson = stats.poisson.pmf(n,rate)
```

```python
poisson[4]
```

```python
1 - (poisson[0]+poisson[1]+poisson[2]+poisson[3])
```

```python
plt.plot(n,poisson,'o-')
plt.show()
```

# Binomial Distribution - 
## Problem Statement 1
A bank has issued credit card to its customers under the scheme of master card. Based on past data, the bank has find out that 60% of all accounts pay on time following the bill. If a sample of 7 accounts is selected at random from the current database, construct the Binomial Probablity Distribution of accounts paying on time.

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = 7
success = 0.6
```

```python
k = np.arange(0,8)
```

```python
binomial = stats.binom.pmf(k,n,success)
binomial
```

```python
plt.plot(k,binomial,'o-')
plt.title("Binomial: n=%i, p=%.2f" %(n,success),fontsize=15)
plt.xlabel("Number of successes")
plt.ylabel("Probablity of successes")
```

```python
cum_binomial = stats.binom.cdf(k,n,success)
cum_binomial
```

## Problem Statement 2
A laptop assembly unit manufactures a single type of Laptop Product and sells it to customers under multiple brand names through its channel partners. From past audit and quality data, it knows that on an average, 10% of all manufactured units since inception has been defective. The manufacturer receives a wholesale order of 100 laptops that are to have highest form of quality and no defective units. However, it is liability of the manufacturer to inform the client about the probablity of a few units being defective out of the whole lot. Construct a Binomial Probablity Distribution of the laptops that are defective out of the 100. 

```python
trials = 100
failure = 0.1
success = 1- failure
success
```

```python
k = np.arange(0,101)
```

```python
binomial = stats.binom.pmf(k,trials,success)
binomial
```

```python
plt.plot(k,binomial,'o-');
```
